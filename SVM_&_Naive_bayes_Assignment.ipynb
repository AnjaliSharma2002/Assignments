```python
# Q1.

# A Support Vector Machine (SVM) is a supervised machine learning algorithm commonly used for classification tasks, though it can also be applied to regression problems. The main objective of an SVM is to find the optimal decision boundary, called a hyperplane, that best separates data points from different classes. It does this by maximizing the margin, which is the distance between the hyperplane and the nearest data points from each class, known as support vectors. These support vectors are crucial because they directly influence the position and orientation of the hyperplane. When the data is not linearly separable, SVM uses a technique called the kernel trick to transform the data into a higher-dimensional space where a linear separation becomes possible. Common kernel functions include the radial basis function (RBF), polynomial, and sigmoid. SVMs are particularly effective in high-dimensional spaces and are used in various applications such as text classification, image recognition, and bioinformatics. However, they may not perform well with large datasets or when the data contains a lot of noise or overlapping classes, and they require careful tuning of parameters and selection of an appropriate kernel.
```

```python
# Q2.

# üîπ Hard Margin SVM:
# Definition: A Hard Margin SVM tries to find a hyperplane that perfectly separates the classes with no misclassifications.

# Assumption: It assumes that the data is linearly separable (i.e., you can draw a line or hyperplane that divides the classes without error).

# Characteristics:

# No tolerance for misclassified points.

# Maximizes the margin without allowing any data points to cross into the wrong side.

# More sensitive to outliers and noise.

# üîπ Soft Margin SVM:
# Definition: A Soft Margin SVM allows some misclassifications to achieve better generalization and to work with non-separable data.

# Assumption: Real-world data is often noisy or not perfectly separable, so it allows some points to violate the margin.

# Characteristics:

# Introduces slack variables to allow margin violations.

# Balances between maximizing the margin and minimizing the classification error.

# Controlled by a regularization parameter (C):

# High C ‚Üí fewer margin violations (behaves more like hard margin).

# Low C ‚Üí allows more margin violations (more flexible to noise).
```

```python
# Q3.

# The mathematical intuition behind SVM is to find the optimal hyperplane that separates data points of different classes with the maximum margin.

# For linearly separable data, SVM solves the optimization problem:

# min 1/2 ‚à•ùë§‚à•^2          subject¬†to¬†yi (w‚ãÖxi+b)‚â•1

# The margin is 2/‚à•w‚à•, and the goal is to maximize it.

# For non-separable data, SVM introduces slack variables and a regularization parameter (C) to allow some misclassifications.

# In non-linear cases, SVM uses kernels to map data into higher dimensions where a linear separator can be found.
```

```python
# Q4.

# In Support Vector Machines (SVM), Lagrange multipliers play a crucial role in solving the constrained optimization problem required to find the optimal hyperplane. The main objective of SVM is to maximize the margin between classes while ensuring that all data points are correctly classified. This involves minimizing a function (the norm of the weight vector) subject to classification constraints. To solve this, the method of Lagrange multipliers is used to construct a Lagrangian function that incorporates both the objective and the constraints. The multipliers, denoted as Œ±i, help enforce these constraints during the optimization process. When this Lagrangian is optimized (in its dual form), it not only simplifies the computation but also makes it possible to use kernel functions for non-linear classification. Importantly, only data points with non-zero Œ±i‚Äã values, called support vectors, directly influence the position of the decision boundary. Thus, Lagrange multipliers are essential in both solving the optimization efficiently and identifying the most critical training samples.
```

```python
# Q5.

# Support vectors in SVM are the data points that lie closest to the decision boundary (hyperplane) and are most critical in determining its position and orientation. These points are the "supporting" elements of the optimal hyperplane because the margin‚Äîthe distance between the classes‚Äîis defined by them.

# In the mathematical formulation of SVM, only the data points that have non-zero Lagrange multipliers (Œ±i>0) are considered support vectors. These points satisfy the condition:

# yi (w‚ãÖxi+b)=1

# Support vectors are important because:

# They directly influence the decision boundary.

# Removing a support vector would change the position of the hyperplane.

# All other data points (far from the boundary) do not affect the hyperplane.

# In both linear and non-linear SVMs (using kernels), support vectors are the key samples that the model uses to make decisions, making SVM a sparse model since it depends only on a subset of the training data.
```

```python
# Q6.

# A Support Vector Classifier (SVC) is the practical implementation of the Support Vector Machine (SVM) algorithm used specifically for classification tasks. It aims to find the optimal hyperplane that separates data points from different classes with the maximum margin, while allowing for some misclassifications in the case of noisy or overlapping data.

# SVC works by solving an optimization problem where it tries to maximize the margin between classes and minimize classification errors. It supports both linear and non-linear classification by using kernel functions (like linear, polynomial, or RBF) to transform the data into higher-dimensional spaces where a linear separator may exist.

# In Python's popular scikit-learn library, SVC is a commonly used class to build SVM models for classification. It includes parameters such as:

# C: controls the trade-off between margin size and classification error (regularization),

# kernel: defines the type of transformation for non-linear classification,

# gamma: defines influence of a single training example in non-linear kernels.
```

```python
# Q7.

# A Support Vector Regressor (SVR) is the regression counterpart of Support Vector Machine (SVM), used for predicting continuous values instead of class labels. Unlike traditional regression methods that try to minimize the error between predicted and actual values, SVR tries to fit the best line (or hyperplane in higher dimensions) within a margin of tolerance, called epsilon (Œµ).

# The key idea of SVR is to ignore errors (differences between predicted and actual values) as long as they are within the Œµ margin. Only the data points that fall outside this margin influence the model‚Äîthese are called the support vectors. The SVR then optimizes the model to be as flat as possible (i.e., minimizing the weight vector norm ‚à•w‚à•^2), while keeping the deviations of predictions within the allowed margin.

# SVR can also use kernel functions (like RBF, polynomial, or linear) to handle non-linear relationships between input and output.
```

```python
# Q8.

# The Kernel Trick in SVM is a clever technique that allows the algorithm to handle non-linearly separable data by implicitly mapping it into a higher-dimensional feature space without explicitly computing the coordinates in that space. Instead of transforming the data points directly (which can be computationally expensive or even impossible), the kernel trick uses a kernel function to compute the inner products of data points in this higher-dimensional space efficiently.

# This way, SVM can find a linear separating hyperplane in the transformed space, which corresponds to a non-linear decision boundary in the original input space. Common kernel functions include:

# Linear kernel ‚Äî no transformation, just the dot product.

# Polynomial kernel ‚Äî maps data into polynomial feature spaces.

# Radial Basis Function (RBF) or Gaussian kernel ‚Äî maps data into infinite-dimensional space, allowing highly flexible boundaries.

# Sigmoid kernel ‚Äî similar to neural networks‚Äô activation function.
```

```python
# Q9.

# The Linear kernel in SVM is the simplest and fastest kernel, computing the direct dot product between input vectors. It works best when the data is linearly separable or in very high-dimensional spaces like text classification, where a linear decision boundary is sufficient. The Polynomial kernel introduces more flexibility by computing a polynomial function of the dot product, allowing SVM to capture moderate non-linear relationships in the data. Its complexity depends on the polynomial degree, and it can model curved decision boundaries, but it‚Äôs more computationally expensive than the linear kernel. The RBF (Radial Basis Function) kernel is the most flexible of the three, measuring similarity based on the distance between points and effectively mapping data into an infinite-dimensional space. This allows it to handle highly complex, non-linear decision boundaries, making it suitable for many real-world problems with intricate patterns. However, it requires careful tuning of parameters like gamma and tends to be computationally heavier. In summary, linear kernels are ideal for simple or very high-dimensional problems, polynomial kernels suit moderately non-linear data, and RBF kernels are preferred for complex non-linear data.
```

```python
# Q10.

# The C parameter in SVM controls the trade-off between maximizing the margin and minimizing classification errors. It acts as a regularization parameter that balances two competing goals: having a wide margin (which promotes better generalization) and correctly classifying all training points (which reduces training errors).

# When C is large, the model tries to classify all training points correctly by penalizing misclassifications heavily. This can lead to a narrower margin and potentially overfitting, where the model fits the training data very tightly but may not generalize well to new data.

# When C is small, the model allows more misclassifications or margin violations to achieve a wider margin. This increases the model‚Äôs flexibility and can improve generalization by being less sensitive to noise, but it may also lead to underfitting if the margin is too wide and ignores important data patterns.
```

```python
# Q11.

# The Gamma (Œ≥) parameter in an SVM with an RBF (Radial Basis Function) kernel controls the influence of a single training example on the shape of the decision boundary. It determines how far the influence of a single data point reaches when creating the boundary.

# When gamma is high, each data point‚Äôs influence is very localized, meaning the decision boundary can become highly flexible and wiggly, closely fitting the training data. This can lead to overfitting, where the model captures noise as if it were a real pattern.

# When gamma is low, the influence of each point is more spread out, resulting in a smoother, less complex decision boundary that generalizes better but might underfit if it‚Äôs too simple to capture the true patterns.
```

```python
# Q12.

# The Na√Øve Bayes classifier is a simple and popular probabilistic machine learning algorithm used for classification tasks. It‚Äôs based on Bayes‚Äô theorem, which calculates the probability of a class given the observed features. The key idea is to predict the class that is most likely to have generated the given data.

# It‚Äôs called ‚Äúna√Øve‚Äù because it makes a strong assumption of feature independence ‚Äî that is, it assumes that all the features in the data are independent of each other given the class label. In reality, this assumption is rarely true, as features often have some level of correlation. Despite this simplification, Na√Øve Bayes often performs surprisingly well, especially in text classification problems like spam detection and sentiment analysis, because the independence assumption simplifies calculations and reduces the amount of training data needed.
```

```python
# Q13.

# Bayes‚Äô Theorem is a fundamental rule in probability theory that describes how to update the probability of a hypothesis based on new evidence. It relates the conditional probabilities of events and is expressed mathematically as:

# P(A‚à£B)= (P(B‚à£A)√óP(A))/P(B)

# Here:

# P(A‚à£B) is the posterior probability ‚Äî the probability of event A happening given that B has occurred.
# P(B‚à£A) is the likelihood ‚Äî the probability of observing event B given that A is true.
# P(A) is the prior probability of event A before observing B.
# P(B) is the marginal probability of event B.

# In simple terms, Bayes‚Äô Theorem helps you update your belief about A when you gain new information B. It‚Äôs widely used in statistics, machine learning, and data science to make probabilistic inferences.
```

```python
# Q14.

# 1. Gaussian Na√Øve Bayes
# Assumption: Features are continuous and follow a Gaussian (normal) distribution.

# Use case: Suitable for datasets where features are continuous-valued (e.g., height, weight, temperature).

# How it works: For each feature, it estimates the mean and variance from the training data and calculates the likelihood using the Gaussian probability density function.

# 2. Multinomial Na√Øve Bayes
# Assumption: Features represent discrete counts or frequencies (e.g., word counts in text documents).

# Use case: Commonly used for text classification and problems involving count data (e.g., document classification, spam detection).

# How it works: Models the probability of features as a multinomial distribution (counts of occurrences), focusing on the frequency of each feature.

# 3. Bernoulli Na√Øve Bayes
# Assumption: Features are binary (0 or 1), indicating presence or absence.

# Use case: Also popular in text classification but focuses on whether a word occurs or not, regardless of frequency.

# How it works: Models features as Bernoulli random variables, calculating probabilities based on the presence or absence of each feature.
```

```python
# Q15.

# You should use Gaussian Na√Øve Bayes when your dataset features are continuous numerical variables that can be reasonably modeled with a normal (Gaussian) distribution. This makes it ideal for problems where the input data consists of measurements like height, weight, temperature, or other real-valued attributes.

# For example, if you have sensor data, medical measurements, or any continuous-valued features where the distribution of each feature roughly resembles a bell curve, Gaussian Na√Øve Bayes is a natural fit. It estimates the mean and variance of each feature for each class to calculate probabilities, which works well under the assumption of normality.

# In contrast, if your features are counts (like word frequencies in documents), the Multinomial Na√Øve Bayes is more suitable; and if your features are binary indicators (presence or absence), then Bernoulli Na√Øve Bayes would be better.

# So, use Gaussian Na√Øve Bayes when your data is continuous and approximately Gaussian-distributed, and you want a simple, fast probabilistic classifier.
```

```python
# Q16.

# The key assumptions made by Na√Øve Bayes are:

# Feature Independence:
# Na√Øve Bayes assumes that all features (predictor variables) are conditionally independent of each other given the class label. In other words, the presence or value of one feature does not affect or depend on another feature within the same class.

# Equal Importance of Features:
# Each feature contributes independently and equally to the final classification decision. There‚Äôs no weighting or interaction considered between features.

# Feature Distribution:
# Depending on the variant (Gaussian, Multinomial, Bernoulli), Na√Øve Bayes assumes a specific probability distribution for the features within each class:

# Gaussian Na√Øve Bayes assumes features follow a normal distribution.

# Multinomial Na√Øve Bayes assumes features follow a multinomial distribution (count data).

# Bernoulli Na√Øve Bayes assumes features are binary (Bernoulli distribution).

# These assumptions simplify the computation of the posterior probabilities, making the model computationally efficient and scalable, even though in real-world data, the independence assumption is often violated. Despite this ‚Äúna√Øve‚Äù assumption, the algorithm performs surprisingly well in many practical scenarios.
```

```python
# Q17.

# Advantages:
# Simple and Fast:
# Na√Øve Bayes is easy to implement and computationally efficient, making it suitable for large datasets and real-time predictions.

# Works Well with High-Dimensional Data:
# It handles large numbers of features well, such as in text classification problems with thousands of words.

# Requires Less Training Data:
# Because of its simplicity and the strong independence assumption, it requires relatively fewer training examples compared to more complex models.

# Performs Well in Practice:
# Despite the strong independence assumption, Na√Øve Bayes often performs surprisingly well, especially for text classification, spam detection, and sentiment analysis.

# Handles Both Continuous and Discrete Data:
# Different variants (Gaussian, Multinomial, Bernoulli) make it versatile across various data types.

# Probabilistic Output:
# It provides probabilistic predictions (class probabilities), which can be useful for decision-making.



# Disadvantages:
# Strong Independence Assumption:
# The assumption that features are conditionally independent is often unrealistic, which can limit performance when features are correlated.

# Zero Probability Problem:
# If a feature value never appears in the training data for a given class, Na√Øve Bayes assigns zero probability to that class (can be mitigated by techniques like Laplace smoothing).

# Not Good with Complex Relationships:
# It struggles when relationships between features are complex or non-linear, since it doesn't model interactions.

# Less Flexible:
# Compared to more advanced models (like Random Forests or Neural Networks), it may have lower accuracy on some complex datasets.
```

```python
# Q18.

# Na√Øve Bayes is a great choice for text classification because of several reasons:

# Handles High-Dimensional Data Efficiently:
# Text data often involves thousands of features (words), and Na√Øve Bayes scales well with this high dimensionality without becoming computationally expensive.

# Works Well with Sparse Data:
# In text, most documents contain only a small subset of all possible words, making the data sparse. Na√Øve Bayes naturally handles such sparse feature vectors.

# Simple and Fast:
# It trains quickly even on large datasets, which is important for practical text applications like spam filtering or sentiment analysis.

# Good Performance Despite Strong Assumptions:
# Although it assumes word independence (which is rarely true), this ‚Äúna√Øve‚Äù assumption often does not hurt performance much in text data, where the presence or absence of words is more important than their correlation.

# Probabilistic Output:
# Na√Øve Bayes provides probabilities of class membership, allowing for confidence scoring and threshold tuning in classification.

# Effective with Limited Training Data:
# It can perform well even when the labeled training data is relatively small, which is common in many text classification tasks.

# Because of these factors, Na√Øve Bayes remains one of the most popular and reliable baseline models for text classification problems.
```

```python
# Q19.

# Support Vector Machines (SVM) and Na√Øve Bayes are both popular classification algorithms but differ fundamentally in their approach and strengths. SVM is a powerful, discriminative model that tries to find the optimal hyperplane to separate classes with maximum margin, making it very effective for both linear and non-linear boundaries (using kernels). It generally performs well on datasets with clear margins and complex boundaries but can be computationally intensive, especially on large datasets. On the other hand, Na√Øve Bayes is a generative probabilistic model that uses Bayes‚Äô theorem with a strong assumption of feature independence, making it much faster and simpler to train. Na√Øve Bayes excels in high-dimensional, sparse datasets like text classification, where its assumptions surprisingly hold well, but it may struggle on datasets with correlated features or complex relationships. While SVM often achieves higher accuracy and is more flexible in modeling complex patterns, Na√Øve Bayes offers speed, scalability, and interpretability, making it a great baseline and choice for certain domains, especially when training data is limited or when probabilistic outputs are needed.
```

```python
# Q20.

# Laplace smoothing (also called add-one smoothing) helps Na√Øve Bayes handle the zero probability problem that occurs when a feature value doesn‚Äôt appear in the training data for a particular class. Without smoothing, if a feature never occurs with a class during training, the model assigns a probability of zero to that feature for that class. Since Na√Øve Bayes multiplies probabilities of features, this zero would make the entire class probability zero, effectively ruling out that class regardless of other evidence.

# Laplace smoothing solves this by adding a small constant (usually 1) to all feature counts before calculating probabilities. This ensures that no feature has zero probability, even if it didn‚Äôt appear in the training set for a class. By doing this, the model becomes more robust, especially when dealing with small or sparse datasets, and improves generalization to unseen data.
```

```python
# Q21.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC()

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM classifier on Iris dataset: {accuracy:.2f}")
```

```python
# Q22.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

wine = datasets.load_wine()
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_linear = SVC(kernel='linear')
svm_rbf = SVC(kernel='rbf')

svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)

y_pred_linear = svm_linear.predict(X_test)
y_pred_rbf = svm_rbf.predict(X_test)

accuracy_linear = accuracy_score(y_test, y_pred_linear)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

print(f"Accuracy with Linear Kernel: {accuracy_linear:.2f}")
print(f"Accuracy with RBF Kernel: {accuracy_rbf:.2f}")
```

```python
# Q24.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

svm_poly = SVC(kernel='poly', degree=3, coef0=1, C=1)
svm_poly.fit(X, y)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)


plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.Paired, edgecolors='k')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("SVM with Polynomial Kernel (degree=3)")
plt.show()
```

```python
# Q26.

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

newsgroups = fetch_20newsgroups(subset='all')

X_train, X_test, y_train, y_test = train_test_split(
    newsgroups.data, newsgroups.target, test_size=0.3, random_state=42)

model = make_pipeline(CountVectorizer(), MultinomialNB())

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Multinomial Na√Øve Bayes on 20 Newsgroups dataset: {accuracy:.2f}")
```

```python
# Q27.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

C_values = [0.1, 1, 10, 100]

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

plt.figure(figsize=(15, 12))

for i, C in enumerate(C_values, 1):
    svm = SVC(kernel='linear', C=C)
    svm.fit(X, y)

    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(2, 2, i)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
    plt.title(f'SVM with Linear Kernel and C={C}')
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')

plt.tight_layout()
plt.show()
```

```python
# Q29.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_unscaled = SVC()
svm_unscaled.fit(X_train, y_train)
y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC()
svm_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Accuracy without scaling: {accuracy_unscaled:.2f}")
print(f"Accuracy with scaling:    {accuracy_scaled:.2f}")
```

```python
# Q30.

# Laplace smoothing (additive smoothing) is mostly relevant for categorical features and Multinomial or Bernoulli Na√Øve Bayes, not typically for Gaussian Na√Øve Bayes, because Gaussian NB models continuous features using a normal distribution and doesn‚Äôt rely on feature counts.

# Python program comparing predictions with and without Laplace smoothing using MultinomialNB

from sklearn.datasets import fetch_20newsgroups_vectorized
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = fetch_20newsgroups_vectorized(subset='train')
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_no_smooth = MultinomialNB(alpha=0)
model_no_smooth.fit(X_train, y_train)
y_pred_no_smooth = model_no_smooth.predict(X_test)
acc_no_smooth = accuracy_score(y_test, y_pred_no_smooth)

model_smooth = MultinomialNB(alpha=1)
model_smooth.fit(X_train, y_train)
y_pred_smooth = model_smooth.predict(X_test)
acc_smooth = accuracy_score(y_test, y_pred_smooth)

print(f"Accuracy without Laplace smoothing (alpha=0): {acc_no_smooth:.4f}")
print(f"Accuracy with Laplace smoothing (alpha=1):    {acc_smooth:.4f}")
```

```python
# Q31.

from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly']
}

grid = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Accuracy: {:.2f}".format(grid.best_score_))

best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)

print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred))
```

```python
# Q33.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

data = {
    'label': ['ham', 'spam', 'ham', 'spam', 'ham'],
    'message': [
        'Hey, are we still meeting today?',
        'Congratulations! You have won a free ticket. Reply WIN.',
        'Don‚Äôt forget to bring the documents.',
        'URGENT! Your mobile number has won a prize.',
        'Let‚Äôs catch up tomorrow for lunch.'
    ]
}
df = pd.DataFrame(data)

df['label'] = df['label'].map({'ham': 0, 'spam': 1})

X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label'], test_size=0.3, random_state=42
)

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)

y_pred = model.predict(X_test_vec)

print("=== Classification Report ===")
print(classification_report(y_test, y_pred))

print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

```python
# Q36.

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score, classification_report

data = load_wine()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

ovr_model = OneVsRestClassifier(SVC(kernel='rbf', C=1.0, gamma='scale'))
ovr_model.fit(X_train_scaled, y_train)
ovr_pred = ovr_model.predict(X_test_scaled)

ovo_model = OneVsOneClassifier(SVC(kernel='rbf', C=1.0, gamma='scale'))
ovo_model.fit(X_train_scaled, y_train)
ovo_pred = ovo_model.predict(X_test_scaled)

print("=== One-vs-Rest (OvR) Strategy ===")
print(classification_report(y_test, ovr_pred))
print(f"OvR Accuracy: {accuracy_score(y_test, ovr_pred):.2f}")

print("\n=== One-vs-One (OvO) Strategy ===")
print(classification_report(y_test, ovo_pred))
print(f"OvO Accuracy: {accuracy_score(y_test, ovo_pred):.2f}")
```

```python
# Q38.

from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import numpy as np

data = load_iris()
X, y = data.data, data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracies = []

for fold, (train_index, test_index) in enumerate(skf.split(X_scaled, y), 1):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = SVC(kernel='rbf', C=1.0, gamma='scale')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Fold {fold} Accuracy: {acc:.2f}")

avg_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy over 5 folds: {avg_accuracy:.2f}")
```

```python
# Q39.

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

data = load_breast_cancer()
X, y = data.data, data.target
class_names = data.target_names

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

prior_configs = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.3, 0.7],
    None
]


for priors in prior_configs:
    print(f"\n=== Prior Probabilities: {priors if priors else 'Learned from data'} ===")

    model = GaussianNB(priors=priors)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.2f}")
    print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
# Q40.

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_full = SVC(kernel='rbf', C=1.0, gamma='scale')
model_full.fit(X_train_scaled, y_train)
y_pred_full = model_full.predict(X_test_scaled)
acc_full = accuracy_score(y_test, y_pred_full)
print(f"Accuracy without RFE: {acc_full:.4f}")

estimator = SVC(kernel='linear')
rfe = RFE(estimator=estimator, n_features_to_select=10)
rfe.fit(X_train_scaled, y_train)

X_train_rfe = rfe.transform(X_train_scaled)
X_test_rfe = rfe.transform(X_test_scaled)

model_rfe = SVC(kernel='rbf', C=1.0, gamma='scale')
model_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = model_rfe.predict(X_test_rfe)
acc_rfe = accuracy_score(y_test, y_pred_rfe)
print(f"Accuracy with RFE (10 features): {acc_rfe:.4f}")
```

```python
# Q45.

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = GaussianNB()
model.fit(X_train_scaled, y_train)

y_proba = model.predict_proba(X_test_scaled)[:, 1]

roc_auc = roc_auc_score(y_test, y_proba)
print(f"ROC-AUC Score: {roc_auc:.4f}")

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'ROC Curve (area = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Gaussian Naive Bayes')
plt.legend(loc='lower right')
plt.show()
```
