```python
# Q1.

# Unsupervised learning is a type of machine learning where the model is not provided with labeled data. Instead, it tries to find patterns, structures, or relationships in the input data on its own.
```

```python
# Q2.

# K-Means clustering is one of the most popular unsupervised learning algorithms used to group data into K number of clusters, where each data point belongs to the cluster with the nearest mean (called the centroid).

# Steps:

# 1. Choose K.

# 2. Randomly initialize K centroids.

# 3. Assign each point to the nearest centroid.

# 4. Update centroids by calculating the mean of assigned points.

# 5. Repeat steps 3–4 until centroids stop changing.
```

```python
# Q3.

# A dendrogram is a tree-like diagram that shows how data points are merged or split step by step in hierarchical clustering.

# - Each leaf represents a data point.

# - Branches show how clusters are combined.

# - The height of a merge shows the distance or dissimilarity between clusters.
```

```python
# Q4.

# The main difference between K-Means and Hierarchical Clustering lies in their approach and flexibility. K-Means is a partition-based method that requires you to specify the number of clusters (K) in advance. It works by assigning data points to the nearest cluster center and updating those centers iteratively. In contrast, Hierarchical Clustering is a tree-based method that builds a hierarchy of clusters without needing to predefine the number of clusters. It produces a dendrogram, which allows you to visualize how clusters are formed and decide the number of clusters later by "cutting" the tree at a certain level. K-Means is generally faster and better suited for large datasets, while Hierarchical Clustering offers more insightful structure but is more computationally intensive, making it ideal for smaller datasets.
```

```python
# Q5.

# - No need to specify the number of clusters (K)
# DBSCAN automatically finds the number of clusters based on data density.

# - Can find clusters of arbitrary shapes
# Unlike K-Means, which assumes spherical clusters, DBSCAN handles irregular shapes well.

# - Handles noise and outliers
# DBSCAN can identify and exclude outliers as noise, while K-Means assigns every point to a cluster.

# - Works better with uneven cluster sizes and densities
# DBSCAN adapts to varying densities, whereas K-Means may struggle in such cases.
```

```python
# Q6.

# When to Use Silhouette Score:

# - To choose the optimal number of clusters (K)
# Try different K values (e.g., in K-Means), and pick the one with the highest average silhouette score.

# - To measure cluster quality
# It tells you how similar a point is to its own cluster compared to other clusters.
```

```python
# Q7.

# - Not scalable for large datasets
# It has high time and space complexity (typically O(n²)), making it slow with thousands of points.

# - No flexibility to revisit decisions
# Once a merge or split is done, it can’t be undone, which might lead to poor clustering.

# - Sensitive to noise and outliers
# Outliers can distort the structure of the hierarchy.

# - Difficulty choosing the right level to cut the dendrogram
# It may be unclear how many clusters to select from the tree.

# - Assumes some distance metric
# Results can vary based on the distance measure (e.g., Euclidean, Manhattan) and linkage method used.
```

```python
# Q8.

# Feature scaling is essential in clustering algorithms like K-Means because these algorithms rely on distance metrics (such as Euclidean distance) to group similar data points. If the features have different scales, such as one being in the range of thousands (e.g., income) and another in the range of single digits (e.g., age), the larger scale feature will disproportionately influence the distance calculation, skewing the results. By scaling the features (using techniques like standardization or min-max normalization), each feature contributes equally to the clustering process. This ensures that the algorithm can form clusters based on all features rather than being biased toward the ones with larger scales, leading to more accurate and meaningful groupings.
```

```python
# Q9.

# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points based on the density of the data. It groups data points into clusters by looking for regions with high density, where points are closely packed together. A point is classified as noise if it does not meet the criteria for being part of any cluster. Specifically, noise points are those that are neither in the core of a cluster (where a sufficient number of neighboring points exist within a specified radius) nor directly reachable from any core point. In DBSCAN, points that do not have enough neighbors to form a dense region are labeled as noise, effectively marking them as outliers. This feature allows DBSCAN to identify and handle noise in the data while focusing on discovering meaningful clusters.
```

```python
# Q10.

# In the context of K-Means clustering, inertia refers to the sum of squared distances between each data point and the centroid of the cluster it belongs to. In simpler terms, it measures how compact the clusters are. A lower inertia indicates that the data points are closer to their respective centroids, meaning the clustering is more efficient.
```

```python
# Q11.

# The Elbow Method in K-Means clustering is a technique used to determine the optimal number of clusters (K). It involves running K-Means clustering on the dataset for a range of K values and then plotting the inertia (sum of squared distances from points to their cluster centroids) against K.

```

```python
# Q12.

# In the context of DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to the concentration of data points in a given area. DBSCAN uses density to identify clusters by grouping together points that are close to each other and have a sufficient number of neighbors within a defined distance. It focuses on finding dense regions of points and separating them from regions with lower density.
```

```python
# Q13.

# Yes, Hierarchical Clustering can be used on categorical data, but it requires a different approach compared to numerical data. The traditional Euclidean distance used in hierarchical clustering is not suitable for categorical data. Instead, you would need to use appropriate distance or similarity measures designed for categorical data.

```

```python
# Q14.

# A negative Silhouette Score indicates that the data points are likely misclassified or incorrectly grouped into clusters. Specifically, it means that a data point is closer to a different cluster than to its own cluster, suggesting poor clustering quality.
```

```python
# Q15.

# The linkage criteria influence how hierarchical clustering merges or splits clusters, affecting the final clustering structure. The choice of linkage impacts the compactness, shape, and size of the resulting clusters, and it's important to choose the appropriate linkage based on the characteristics of your data.
```

```python
# Q16.

# K-Means clustering can perform poorly on data with varying cluster sizes or densities because of its inherent assumptions and limitations. The algorithm assumes that clusters are spherical and have roughly equal sizes, meaning it struggles when clusters have different shapes, such as elongated or irregular forms. It is also sensitive to the initial selection of centroids; if the centroids are poorly chosen, K-Means may assign points to the wrong clusters, especially in data with clusters of varying densities. In such cases, the algorithm tends to favor larger clusters and may fail to correctly identify smaller or denser clusters. Additionally, K-Means uses a hard assignment approach, where each point is assigned to one cluster, which can be problematic when clusters overlap or vary in density. As a result, K-Means may incorrectly merge or split clusters in datasets where the clusters have unequal sizes or densities, leading to poor clustering performance.
```

```python
# Q17.

# In DBSCAN, the two core parameters are ε (epsilon) and MinPts.

# ε (epsilon): Defines the radius around each point to determine which points are considered neighbors. A small ε results in many small clusters or noise, while a large ε can merge clusters together.

# MinPts: Specifies the minimum number of points required to form a dense region (cluster). A higher MinPts value creates fewer, larger clusters, while a lower value may result in smaller clusters or overfitting.

# Together, these parameters control how DBSCAN forms clusters and handles noise, and their settings must align with the dataset’s density and structure.
```

```python
# Q18.

# K-Means++ improves upon standard K-Means by smartly selecting initial centroids. Instead of choosing them randomly, it selects the first centroid randomly and then picks subsequent centroids based on their distance from the already chosen ones. This ensures the centroids are spread out, reducing the chances of poor initialization. As a result, K-Means++ leads to faster convergence and better clustering quality compared to the standard K-Means algorithm.
```

```python
# Q19.

# Agglomerative clustering is a type of hierarchical clustering where the algorithm starts with each data point as its own individual cluster and then iteratively merges the closest clusters based on a defined similarity or distance measure until all points are grouped into a single cluster.
```

```python
# Q20.

# The Silhouette Score is a better metric than inertia because it evaluates both compactness (how close points are to their cluster centroid) and separation (how distinct clusters are from each other). In contrast, inertia only measures compactness, which can be misleading if clusters overlap. The Silhouette Score provides a more comprehensive and interpretable evaluation of clustering quality, especially when comparing different cluster configurations.
```

```python
# Q23.

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)

dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(8, 6))
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title("DBSCAN Clustering on make_moons Data\n(Black dots = Outliers)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Q29.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=1000, centers=4, cluster_std=1.0, random_state=42)

silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering (k = 2 to 5)')
plt.xticks(k_values)
plt.grid(True, axis='y')
plt.show()
```

```python
# Q34.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

data = load_breast_cancer()
X = data.data
y = data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertia_values = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia_values, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K (Breast Cancer Dataset)')
plt.xticks(k_values)
plt.grid(True)
plt.show()

print("Inertia values for K = 2 to 6:")
for k, inertia in zip(k_values, inertia_values):
    print(f"K = {k}: Inertia = {inertia:.2f}")
```

```python
# Q36.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from collections import Counter

wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=1.0, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

unique_clusters = np.unique(clusters)
n_clusters = len(unique_clusters) - 1
n_noise = np.sum(clusters == -1)

print("Cluster assignments:", Counter(clusters))
print(f"\nNumber of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {n_noise}")
print(f"Total data points: {len(X)}")

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.xlabel(f"{feature_names[0]} (scaled)")
plt.ylabel(f"{feature_names[1]} (scaled)")
plt.title(f"DBSCAN Clustering on Wine Dataset\n{len(unique_clusters)-1} clusters found (excluding noise)")
plt.grid(True)
plt.show()
```

```python
# Q38.

import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

iris = load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

noise_count = np.sum(clusters == -1)
total_samples = len(X)

print("DBSCAN Cluster Assignments:", np.unique(clusters, return_counts=True))
print(f"\nNumber of noise points: {noise_count} ({(noise_count/total_samples)*100:.1f}% of total)")
print(f"Number of clusters found: {len(set(clusters)) - (1 if -1 in clusters else 0)}")

print("\nComparison with actual species:")
for cluster_num in np.unique(clusters):
    cluster_mask = (clusters == cluster_num)
    species_counts = np.bincount(y[cluster_mask])
    print(f"Cluster {cluster_num}: {species_counts}")
```

```python
# Q39.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

X, y = make_moons(n_samples=300, noise=0.08, random_state=42)

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

centers = kmeans.cluster_centers_

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50)
plt.title("Original Moon Data (Ground Truth)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', alpha=0.8, edgecolor='black')
plt.title("K-Means Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()

from sklearn.metrics import adjusted_rand_score, silhouette_score
print(f"Adjusted Rand Index: {adjusted_rand_score(y, y_kmeans):.3f}")
print(f"Silhouette Score: {silhouette_score(X, y_kmeans):.3f}")
```

```python
# Q40.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

digits = load_digits()
X = digits.data

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X_pca)
labels = kmeans.labels_

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels, cmap='tab10', marker='o')

ax.set_title('KMeans Clustering on Digits Dataset (PCA Reduced to 3D)')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')

plt.show()
```

```python
# Q41.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=1000, centers=5, random_state=42)

kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(X)

score = silhouette_score(X, labels)
print(f"Silhouette Score: {score:.3f}")

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("KMeans Clustering of Synthetic Blobs")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Q42.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

data = load_breast_cancer()
X = data.data

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

agg_clustering = AgglomerativeClustering(n_clusters=2)
labels = agg_clustering.fit_predict(X_pca)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("Agglomerative Clustering on PCA-reduced Breast Cancer Dataset")
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Q43.

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN

X, _ = make_circles(n_samples=1000, noise=0.1, factor=0.5, random_state=42)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
plt.title("KMeans Clustering")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
# Q44.

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import numpy as np

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

silhouette_vals = silhouette_samples(X, labels)

plt.figure(figsize=(8, 6))
plt.scatter(range(len(silhouette_vals)), silhouette_vals, c=labels, cmap='viridis')
plt.title("Silhouette Coefficients for each sample after KMeans Clustering")
plt.xlabel('Sample Index')
plt.ylabel('Silhouette Coefficient')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Q45.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

X, _ = make_blobs(n_samples=1000, centers=4, random_state=42)

agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='average')
labels = agg_clustering.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("Agglomerative Clustering with 'Average' Linkage")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Q46.

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

wine = load_wine()
X = wine.data[:, :4]
features = wine.feature_names[:4]

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

df = pd.DataFrame(X, columns=features)
df['Cluster'] = labels

sns.pairplot(df, hue='Cluster', palette='viridis')
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
# Q48.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

agg_clustering = AgglomerativeClustering(n_clusters=10)
labels = agg_clustering.fit_predict(X_tsne)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("Agglomerative Clustering on t-SNE Reduced Digits Dataset")
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
