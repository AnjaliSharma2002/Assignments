```python
# Q1.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
x = np.random.rand(100) * 10
y = 2.5 * x + np.random.randn(100) * 3

sns.set_style("whitegrid")

plt.figure(figsize=(7, 5))
sns.scatterplot(x=x, y=y, color='blue', edgecolor='black', alpha=0.7)

plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.title("Scatter Plot of Synthetic Data")

plt.show()
```

```python
# Q2.


np.random.seed(42)
data = np.random.randn(1000)

sns.set_style("whitegrid")

plt.figure(figsize=(7, 5))
sns.histplot(data, bins=20, kde=True, color='blue', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data with Seaborn")

plt.show()

```

```python
# PLOTLY ASSIGNMENT:

import plotly.graph_objects as go
import plotly.express as px
```

```python
# Q2.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(
    df,
    x="Grade",
    y="Score",
    box=True,
    points="all",
    color="Grade",
    title="Distribution of Student Scores Across Grades"
)

fig.show()

```

```python

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = px.imshow(
    sales_pivot,
    labels={'x': 'Month', 'y': 'Day', 'color': 'Sales'},
    x=sales_pivot.columns,
    y=sales_pivot.index,
    color_continuous_scale='Viridis'
)

fig.update_layout(
    title="Sales Heatmap (Variation Across Months & Days)",
    xaxis_title="Month",
    yaxis_title="Day"
)

fig.show()
```

```python
# Q4.

import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

fig.update_layout(
    title="3D Surface Plot of z = sin(sqrt(x + y))",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

fig.show()
```

```python
# Q5.

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = px.scatter(
    df,
    x="GDP",
    y="Population",
    size="Population",
    text="Country",
    title="Bubble Chart of Countries: GDP vs Population",
    labels={"GDP": "Gross Domestic Product (Billion $)", "Population": "Population (Millions)"},
    color="Country",
    hover_name="Country"
)

fig.show()
```

```python
# Q1.

from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave Plot", x_axis_label='X', y_axis_label='sin(X)')
p.line(x, y, line_width=2, color="blue")

show(p)
```

```python
# Q2.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.randint(10, 50, size=50)
colors = np.random.choice(["red", "blue", "green", "orange"], 50)

source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

show(p)
```

```python
# Q3.

from bokeh.plotting import figure, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Count Bar Chart", x_axis_label='Fruit', y_axis_label='Count')
p.vbar(x=fruits, top=counts, width=0.5, color=['red', 'orange', 'yellow', 'green'])

show(p)
```

```python
# Q4.

from bokeh.plotting import figure, show
import numpy as np

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="black")

show(p)
```

```python
# Q5.

from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
import numpy as np

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

show(p)
```
